```python

try:
    import w3lib
except ImportError:
    import subprocess
    subprocess.run(['pip', 'install', '-q', 'w3lib'], check=True)

import sys
sys.path.append('/content/drive/MyDrive/vip2025/DeepSC-master')

import torch
import json
from models.transceiver import DeepSC
from dataset import EurDataset
from utils import initNetParams, SNR_to_noise
import torch.nn.functional as F
import gc
import numpy as np

# ============================================================================
# SETTINGS
# ============================================================================
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

K = 3           # Number of models: 3, 5, or 7
SNR = 0         # SNR in dB
NUM_SAMPLES = 100  # Start with 100 samples

# Model params
channel = 'Rayleigh'
max_length = 30
d_model = 128
num_heads = 8
dff = 512

print(f"EXTREME Memory-Safe Mode")
print(f"  K = {K} models")
print(f"  SNR = {SNR} dB")
print(f"  Samples = {NUM_SAMPLES}")
print(f"  Processing: 1 sample at a time\n")

# ============================================================================
# Load vocabulary
# ============================================================================
with open('/content/drive/MyDrive/vip2025/europarl/vocab.json', 'r') as f:
    vocab = json.load(f)

if isinstance(vocab, dict):
    if 'token_to_idx' in vocab:
        vocab_size = len(vocab['token_to_idx'])
        idx_to_token = vocab.get('idx_to_token') or {str(v): k for k, v in vocab['token_to_idx'].items()}
    else:
        vocab_size = len(vocab)
        idx_to_token = {str(v): k for k, v in vocab.items()}

print(f"Vocab size: {vocab_size}\n")

# ============================================================================
# Load test dataset - NO DataLoader, direct access
# ============================================================================
test_dataset = EurDataset('test')
noise_std = SNR_to_noise(SNR)

print(f"Test dataset loaded: {len(test_dataset)} samples\n")

# ============================================================================
# Simple accuracy metric
# ============================================================================
def word_accuracy(pred, tgt):
    """Simple word-level accuracy"""
    # Handle both tensor and list inputs
    if torch.is_tensor(pred):
        pred = pred.cpu().numpy()
    if torch.is_tensor(tgt):
        tgt = tgt.cpu().numpy()

    pred_clean = [int(idx) for idx in pred if int(idx) not in [0, 1, 2, 3]]
    tgt_clean = [int(idx) for idx in tgt if int(idx) not in [0, 1, 2, 3]]

    min_len = min(len(pred_clean), len(tgt_clean))
    if min_len == 0:
        return 0.0

    correct = sum([p == t for p, t in zip(pred_clean[:min_len], tgt_clean[:min_len])])
    return correct / min_len

# ============================================================================
# Inline fusion functions
# ============================================================================
def compute_entropy(logits):
    probs = F.softmax(logits, dim=-1)
    log_probs = torch.log(probs + 1e-10)
    entropy = -torch.sum(probs * log_probs, dim=-1)
    return entropy / np.log(logits.shape[-1])

def confidence_weighted_fusion(logits_list):
    stacked = torch.stack(logits_list, dim=0)
    entropies = torch.stack([compute_entropy(l) for l in logits_list], dim=0)
    confidences = 1.0 - entropies
    weights = confidences / (confidences.sum(dim=0, keepdim=True) + 1e-10)
    return (weights.unsqueeze(-1) * stacked).sum(dim=0)

def top_k_fusion(logits_list, k):
    K = len(logits_list)
    k = min(k, K)
    stacked = torch.stack(logits_list, dim=0)
    entropies = torch.stack([compute_entropy(l) for l in logits_list], dim=0)
    confidences = 1.0 - entropies
    _, top_k_idx = torch.topk(confidences.permute(1, 2, 0), k=k, dim=-1)

    vocab_size = logits_list[0].shape[-1]
    top_k_idx_exp = top_k_idx.unsqueeze(-1).expand(-1, -1, -1, vocab_size)
    stacked_perm = stacked.permute(1, 2, 0, 3)
    selected = torch.gather(stacked_perm, dim=2, index=top_k_idx_exp)

    conf_perm = confidences.permute(1, 2, 0)
    sel_conf = torch.gather(conf_perm, dim=2, index=top_k_idx)
    weights = sel_conf / (sel_conf.sum(dim=2, keepdim=True) + 1e-10)
    return (weights.unsqueeze(-1) * selected).sum(dim=2)

# ============================================================================
# MAIN EVALUATION LOOP
# ============================================================================
BASE_CKPT = "/content/drive/MyDrive/vip2025/europarl/checkpoints/deepsc-Rayleigh"

results = {
    'simple_average': [],
    'confidence_weighted': [],
    'top1': [],
    'top2': [],
    'top3': [],
}

print(f"Processing {NUM_SAMPLES} samples (one at a time)...\n")

for sample_idx in range(NUM_SAMPLES):
    if sample_idx % 20 == 0:
        print(f"Sample {sample_idx}/{NUM_SAMPLES}...")

    # Get sample directly from dataset
    sample = test_dataset[sample_idx]

    # Handle different dataset return formats
    if isinstance(sample, (list, tuple)):
        if len(sample) >= 2:
            src_data = sample[0]
            tgt_data = sample[1]
        else:
            print(f"ERROR: Sample has unexpected length: {len(sample)}")
            continue
    else:
        print(f"ERROR: Sample has unexpected type: {type(sample)}")
        continue

    # Convert to tensors if needed
    if not torch.is_tensor(src_data):
        src_data = torch.tensor(src_data)
    if not torch.is_tensor(tgt_data):
        tgt_data = torch.tensor(tgt_data)

    # Add batch dimension and move to device
    src = src_data.unsqueeze(0).to(device)  # Shape: (1, seq_len)
    tgt = tgt_data  # Keep on CPU

    # Get predictions from all K models
    model_logits = []

    for seed in range(K):
        ckpt_path = f"{BASE_CKPT}/k{K}/seed{seed}/best_model.pth"

        # Load model
        model = DeepSC(vocab_size, vocab_size, channel, max_length, max_length,
                      d_model, num_heads, dff)
        model = initNetParams(model)
        model.load_state_dict(torch.load(ckpt_path, map_location=device))
        model.to(device)
        model.eval()

        # Get prediction
        with torch.no_grad():
            logits = model(src, max_length, noise_std)
            model_logits.append(logits)

        # Delete model immediately
        del model
        torch.cuda.empty_cache()

    # Evaluate all fusion methods
    # Method 1: Simple Average
    avg_logits = torch.stack(model_logits, dim=0).mean(dim=0)
    avg_pred = torch.argmax(avg_logits, dim=-1).squeeze(0)
    results['simple_average'].append(word_accuracy(avg_pred, tgt))

    # Method 2: Confidence-Weighted
    cw_logits = confidence_weighted_fusion(model_logits)
    cw_pred = torch.argmax(cw_logits, dim=-1).squeeze(0)
    results['confidence_weighted'].append(word_accuracy(cw_pred, tgt))

    # Method 3-5: Top-K
    for k_val, name in [(1, 'top1'), (2, 'top2'), (3, 'top3')]:
        k_use = min(k_val, K)
        topk_logits = top_k_fusion(model_logits, k=k_use)
        topk_pred = torch.argmax(topk_logits, dim=-1).squeeze(0)
        results[name].append(word_accuracy(topk_pred, tgt))

    # Clear everything
    del model_logits, avg_logits, cw_logits, topk_logits, src, tgt
    torch.cuda.empty_cache()
    gc.collect()

# ============================================================================
# Print Results
# ============================================================================
print(f"\n{'='*75}")
print(f"{'RESULTS':^75}")
print(f"K={K} models, SNR={SNR}dB, {NUM_SAMPLES} samples")
print(f"{'='*75}\n")

print(f"{'Method':<25} {'Score':>10} {'Improvement':>15}")
print(f"{'-'*75}")

baseline = np.mean(results['simple_average'])
for method, scores in results.items():
    score = np.mean(scores)
    improvement = ((score - baseline) / baseline) * 100 if baseline > 0 else 0

    method_name = method.replace('_', ' ').title()
    if score == max([np.mean(v) for v in results.values()]):
        method_name += " ⭐"

    baseline_text = "(baseline)" if method == 'simple_average' else f"{improvement:+.2f}%"
    print(f"{method_name:<25} {score:>10.4f} {baseline_text:>15}")

print(f"{'='*75}\n")

best = max(results.items(), key=lambda x: np.mean(x[1]))
print(f"🏆 Best: {best[0]} (score: {np.mean(best[1]):.4f})\n")

print(f"✅ Complete! Tested {NUM_SAMPLES} samples without crashing.")
print(f"💡 Increase NUM_SAMPLES to 200, then 500 if this worked.\n")
```

```python
import argparse
import json
import os
import math
from pathlib import Path

import numpy as np
import torch
from torch.utils.data import DataLoader
from tqdm import tqdm
from sentence_transformers import SentenceTransformer, util

from dataset import EurDataset, collate_data
from models.transceiver import DeepSC
from utils import BleuScore, SNR_to_noise, SeqtoText, subsequent_mask, PowerNormalize, Channels

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")


def list_checkpoints(base_dir, k):
    paths = []
    for seed_dir in sorted(Path(base_dir).glob("seed*")):
        ckpts = sorted(seed_dir.glob("checkpoint_*.pth"))
        if ckpts:
            paths.append(str(ckpts[-1]))
    if len(paths) < k:
        raise RuntimeError(f"Found only {len(paths)} checkpoints under {base_dir}, need {k}")
    return paths[:k]


def load_models(ckpt_paths, cfg):
    models = []
    for p in ckpt_paths:
        model = DeepSC(**cfg).to(device)
        state = torch.load(p, map_location=device)
        model.load_state_dict(state)
        model.eval()
        models.append(model)
    return models


def forward_channel(model, src, noise_var, channel):
    channels = Channels()
    pad_idx = cfg_global["pad"]
    src_mask = (src == pad_idx).unsqueeze(-2).float().to(device)
    enc = model.encoder(src, src_mask)
    tx = PowerNormalize(model.channel_encoder(enc))
    if channel == "AWGN":
        rx = channels.AWGN(tx, noise_var)
    elif channel == "Rayleigh":
        rx = channels.Rayleigh(tx, noise_var)
    else:
        rx = channels.Rician(tx, noise_var)
    memory = model.channel_decoder(rx)
    return memory


def step_logits(model, memory, outputs):
    pad_idx = cfg_global["pad"]
    trg_mask = (outputs == pad_idx).unsqueeze(-2).float()
    look_ahead = subsequent_mask(outputs.size(1)).type_as(trg_mask)
    mask = torch.max(trg_mask, look_ahead).to(device)
    dec = model.decoder(outputs, memory, mask, None)
    return model.dense(dec)


def fuse_logits(logits_list, mode):
    stacked = torch.stack(logits_list, dim=0)  # [K,B,L,V]
    if mode == "uniform":
        return stacked.mean(dim=0)
    probs = torch.softmax(stacked, dim=-1)
    ent = -(probs * torch.log(probs + 1e-10)).sum(dim=-1)
    norm_ent = ent / math.log(probs.size(-1))
    confidences = 1 - norm_ent
    if mode.startswith("ttopk"):
        k = int(mode.replace("ttopk", "") or 1)
        k = min(k, probs.size(0))
        conf_perm = confidences.permute(1, 2, 0)  # [B,L,K]
        _, idx = torch.topk(conf_perm, k=k, dim=-1)
        probs_perm = probs.permute(1, 2, 0, 3)
        idx_exp = idx.unsqueeze(-1).expand(-1, -1, -1, probs.size(-1))
        top_probs = torch.gather(probs_perm, dim=2, index=idx_exp)
        weights = torch.ones_like(idx, dtype=probs.dtype) / k
        fused = (weights.unsqueeze(-1) * top_probs).sum(dim=2)
        return fused
    if mode.startswith("topkcwa"):
        k = int(mode.replace("topkcwa", "") or 2)
        k = min(k, probs.size(0))

        # Compute confidence as before
        probs = torch.softmax(stacked, dim=-1)
        ent = -(probs * torch.log(probs + 1e-10)).sum(dim=-1)
        norm_ent = ent / math.log(probs.size(-1))
        conf = 1 - norm_ent  # [K, B, L]

        # Select top-k models
        conf_perm = conf.permute(1, 2, 0)        # [B, L, K]
        top_conf, idx = torch.topk(conf_perm, k=k, dim=-1)

        probs_perm = probs.permute(1, 2, 0, 3)   # [B, L, K, V]
        idx_exp = idx.unsqueeze(-1).expand(-1, -1, -1, probs.size(-1))
        top_probs = torch.gather(probs_perm, dim=2, index=idx_exp)

        # Normalize top-k confidences
        weights = top_conf / (top_conf.sum(dim=-1, keepdim=True) + 1e-10)
        weights = weights.unsqueeze(-1)

        # Weighted fusion
        fused = (weights * top_probs).sum(dim=2)
        return fused

    # default confidence-weighted
    weights = confidences / (confidences.sum(dim=0, keepdim=True) + 1e-10)
    fused = (weights.unsqueeze(-1) * probs).sum(dim=0)
    return fused


@torch.no_grad()
def decode_with_fusion(models, src, noise_var, max_len, mode):
    memories = [forward_channel(m, src, noise_var, cfg_global["channel"]) for m in models]
    outputs = torch.ones(src.size(0), 1).fill_(cfg_global["start"]).type_as(src)
    for _ in range(max_len - 1):
        logits = [step_logits(m, mem, outputs)[:, -1:, :] for m, mem in zip(models, memories)]
        fused_probs = fuse_logits(logits, mode)
        next_word = torch.argmax(fused_probs, dim=-1)
        outputs = torch.cat([outputs, next_word], dim=1)
    return outputs


def evaluate(args):
    ckpts = list_checkpoints(args.ckpt_root, args.k)
    print(f"Loaded checkpoints:\n" + "\n".join(f"  {p}" for p in ckpts))
    models = load_models(ckpts, cfg_global["model_cfg"])
    dataset = EurDataset(split="test")
    loader = DataLoader(dataset, batch_size=args.batch_size, num_workers=0,
                        pin_memory=True, collate_fn=collate_data)
    bleu = BleuScore(1, 0, 0, 0)
    sbert = SentenceTransformer("all-MiniLM-L6-v2")
    seq2text = SeqtoText(cfg_global["tok2idx"], cfg_global["end"])
    metrics = {"snr": [], "bleu": [], "sim": []}
    for snr in args.snrs:
        noise = SNR_to_noise(snr)
        tx, rx = [], []
        batches = 0
        for batch in tqdm(loader, desc=f"SNR {snr}"):
            batch = batch.to(device)
            preds = decode_with_fusion(models, batch, noise, args.max_len, args.fusion)
            tx.extend(map(seq2text.sequence_to_text, preds.cpu().numpy().tolist()))
            rx.extend(map(seq2text.sequence_to_text, batch.cpu().numpy().tolist()))
            batches += 1
            if args.max_batches and batches >= args.max_batches:
                break
        bleu_scores = bleu.compute_blue_score(rx, tx)
        sim_scores = sbert.encode(rx, convert_to_tensor=True)
        pred_embed = sbert.encode(tx, convert_to_tensor=True)
        sim_val = util.cos_sim(sim_scores, pred_embed).diagonal().cpu().numpy()
        metrics["snr"].append(snr)
        metrics["bleu"].append(float(np.mean(bleu_scores)))
        metrics["sim"].append(float(np.mean(sim_val)))
        print(f"SNR {snr}dB -> BLEU {metrics['bleu'][-1]:.4f}, Sim {metrics['sim'][-1]:.4f}")
    os.makedirs(args.output_dir, exist_ok=True)
    out_file = Path(args.output_dir) / f"ensemble_k{args.k}_{args.fusion}.json"
    with open(out_file, "w", encoding="utf-8") as f:
        json.dump(metrics, f, indent=2)
    print(f"Saved metrics to {out_file}")


def parse_args():
    parser = argparse.ArgumentParser()
    parser.add_argument("--ckpt-root", default = "../europarl/checkpoints/deepsc-Rayleigh/k5_jammoul/rayleigh_channel/")
    parser.add_argument("--fusion", default="topkcwa5")
    parser.add_argument("--snrs", default="0,6,12,18", help="Comma separated SNR list")
    parser.add_argument("--k", type=int, default=5)
    parser.add_argument("--batch-size", type=int, default=32)
    parser.add_argument("--max-len", type=int, default=30)
    parser.add_argument("--max-batches", type=int, default=0)
    parser.add_argument("--output-dir", default="../europarl/checkpoints/deepsc-Rayleigh/k5_jammoul/rayleigh_channel/results_cwa/")
    parser.add_argument("--config", help="Optional JSON config for model architecture")
    return parser.parse_args(args=[])


if __name__ == "__main__":
    args = parse_args()
    args.snrs = [int(x) for x in args.snrs.split(",") if x]
    vocab = json.load(open(os.path.join("..", "europarl", "vocab.json"), "rb")) # Adjusted path to vocab.json
    if args.config:
        with open(args.config, "r", encoding="utf-8") as f:
            model_cfg = json.load(f)
    else:
        model_cfg = dict(
            num_layers=4,
            src_vocab_size=len(vocab["token_to_idx"]),
            trg_vocab_size=len(vocab["token_to_idx"]),
            src_max_len=len(vocab["token_to_idx"]),
            trg_max_len=len(vocab["token_to_idx"]),
            d_model=128,
            num_heads=8,
            dff=512,
            dropout=0.1,
        )
    model_cfg["src_vocab_size"] = len(vocab["token_to_idx"])
    model_cfg["trg_vocab_size"] = len(vocab["token_to_idx"])
    model_cfg["src_max_len"] = len(vocab["token_to_idx"])
    model_cfg["trg_max_len"] = len(vocab["token_to_idx"])
    cfg_global = {
        "tok2idx": vocab["token_to_idx"],
        "pad": vocab["token_to_idx"]["<PAD>"],
        "start": vocab["token_to_idx"]["<START>"],
        "end": vocab["token_to_idx"]["<END>"],
        "channel": "Rayleigh",
        "model_cfg": model_cfg,
    }
    evaluate(args)
```

```python

# ===================================================================
# COMPLETE ALL-IN-ONE KD COMPARISON - FIXED & ENHANCED
# No subprocess, includes training, evaluation, CSV, and plots
# ===================================================================

import os
import json
import torch
import torch.nn as nn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
from tqdm.notebook import tqdm
import copy
import random
import math
from sentence_transformers import SentenceTransformer, util

# Your imports
from dataset import EurDataset, collate_data
from models.transceiver import DeepSC
from utils import BleuScore, SNR_to_noise, SeqtoText, initNetParams, PowerNormalize, Channels, subsequent_mask
from torch.utils.data import DataLoader

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# ===================================================================
# CONFIGURATION
# ===================================================================

# ===================================================================
# CONFIGURATION
# ===================================================================

CONFIG = {
    "vocab_file": "../europarl/vocab.json",
    "teacher_ckpt_root": "../europarl/checkpoints/deepsc-Rayleigh/k5_jammoul/rayleigh_channel/",
    "channel": "Rayleigh",
    "epochs": 40,
    "batch_size": 64,
    "k_teachers": 5,
    "temperature": 2.0,
    "alpha": 0.9, # TUNED: High alpha to trust teacher and prevent overfitting
    "feature_loss_weight": 0.1,
    "seed": 42,
    "student_d_model": 128,
    "student_dff": 256,
    "student_num_layers": 2,
    "student_num_heads": 8,
    "teacher_d_model": 128,
    "teacher_dff": 512,
    "teacher_num_layers": 4,
    "teacher_num_heads": 8,
}

KD_METHODS = [
    {
        'name': 'hinton',
        'description': 'Hinton Temperature Distillation (uniform averaging)',
        'paper': 'Hinton et al. 2015',
        'params': {}
    },
    {
        'name': 'cwa',
        'description': 'Confidence-Weighted Averaging (your method)',
        'paper': 'Your ensemble work',
        'params': {}
    },
    {
        'name': 'ueed',
        'description': 'UEED - Correctness-weighted teachers',
        'paper': 'Wu et al. 2022',
        'params': {}
    },
    {
        'name': 'robust_mean',
        'description': 'Robust Aggregation - Mean of logits',
        'paper': 'Hamdi et al. 2025',
        'params': {}
    },
    {
        'name': 'robust_median',
        'description': 'Robust Aggregation - Median',
        'paper': 'Hamdi et al. 2025',
        'params': {}
    },
    {
        'name': 'robust_trimmed',
        'description': 'Robust Aggregation - Trimmed mean',
        'paper': 'Hamdi et al. 2025',
        'params': {'robust_trim_ratio': 0.2}
    },
    {
        'name': 'robust_krum',
        'description': 'Robust Aggregation - Krum (consensus)',
        'paper': 'Hamdi et al. 2025',
        'params': {}
    },
]

# ===================================================================
# HELPER FUNCTIONS
# ===================================================================

def setup_seed(seed):
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    np.random.seed(seed)
    random.seed(seed)
    torch.backends.cudnn.deterministic = True

def list_teacher_checkpoints(base_dir, k):
    paths = []
    for seed_dir in sorted(Path(base_dir).glob("seed*")):
        ckpts = sorted(seed_dir.glob("checkpoint_*.pth"))
        if ckpts:
            paths.append(str(ckpts[-1]))
    if len(paths) < k:
        raise RuntimeError(f"Found only {len(paths)} checkpoints under {base_dir}, need {k}")
    return paths[:k]

def load_teacher_models(ckpt_paths, num_vocab, config):
    models = []
    for i, p in enumerate(ckpt_paths):
        print(f"Loading teacher {i+1}/{len(ckpt_paths)}...", flush=True)
        model = DeepSC(
            config["teacher_num_layers"], num_vocab, num_vocab,
            num_vocab, num_vocab, config["teacher_d_model"],
            config["teacher_num_heads"], config["teacher_dff"], 0.1
        ).to(device)
        state = torch.load(p, map_location=device)
        model.load_state_dict(state)
        model.eval()
        models.append(model)
    return models

# ===================================================================
# FUSION METHODS
# ===================================================================

def fusion_hinton(stacked_logits, temperature, **kwargs):
    probs_temp = torch.softmax(stacked_logits / temperature, dim=-1)
    return probs_temp.mean(dim=0)

def fusion_cwa(stacked_logits, temperature, **kwargs):
    probs = torch.softmax(stacked_logits, dim=-1)
    ent = -(probs * torch.log(probs + 1e-10)).sum(dim=-1)
    norm_ent = ent / math.log(probs.size(-1))
    confidences = 1 - norm_ent
    weights = confidences / (confidences.sum(dim=0, keepdim=True) + 1e-10)
    weights = weights.unsqueeze(-1)
    probs_temp = torch.softmax(stacked_logits / temperature, dim=-1)
    return (weights * probs_temp).sum(dim=0)

def fusion_ueed(stacked_logits, temperature, true_labels=None, pad_idx=None, **kwargs):
    probs = torch.softmax(stacked_logits, dim=-1)
    if true_labels is not None:
        teacher_preds = torch.argmax(probs, dim=-1)
        true_labels_exp = true_labels.unsqueeze(0).expand(probs.size(0), -1, -1)
        correct = (teacher_preds == true_labels_exp).float()
        if pad_idx is not None:
            pad_mask = (true_labels != pad_idx).float()
            correct = correct * pad_mask.unsqueeze(0)
        weights = correct / (correct.sum(dim=0, keepdim=True) + 1e-10)
        weights = weights.unsqueeze(-1)
        probs_temp = torch.softmax(stacked_logits / temperature, dim=-1)
        return (weights * probs_temp).sum(dim=0)
    else:
        probs_temp = torch.softmax(stacked_logits / temperature, dim=-1)
        return probs_temp.mean(dim=0)

def fusion_robust_median(stacked_logits, temperature, **kwargs):
    median_logits = torch.median(stacked_logits, dim=0).values
    return torch.softmax(median_logits / temperature, dim=-1)

def fusion_robust_trimmed_mean(stacked_logits, temperature, trim_ratio=0.2, **kwargs):
    K = stacked_logits.size(0)
    k_trim = max(1, int(K * trim_ratio))
    sorted_logits, _ = torch.sort(stacked_logits, dim=0)
    trimmed_logits = sorted_logits[k_trim:-k_trim] if k_trim > 0 else sorted_logits
    mean_logits = trimmed_logits.mean(dim=0)
    return torch.softmax(mean_logits / temperature, dim=-1)

def fusion_robust_krum(stacked_logits, temperature, **kwargs):
    K = stacked_logits.shape[0]
    flat_logits = stacked_logits.reshape(K, -1)
    distances = torch.zeros(K, device=stacked_logits.device)
    for k in range(K):
        for j in range(K):
            if j != k:
                distances[k] += torch.sum((flat_logits[k] - flat_logits[j]) ** 2)
    best_teacher_idx = torch.argmin(distances)
    consensus_logits = stacked_logits[best_teacher_idx]
    return torch.softmax(consensus_logits / temperature, dim=-1)

def fusion_robust_mean(stacked_logits, temperature, **kwargs):
    mean_logits = stacked_logits.mean(dim=0)
    return torch.softmax(mean_logits / temperature, dim=-1)

# ===================================================================
# UNIFIED TARGET GENERATION & DATA RETRIEVAL
# ===================================================================

def get_teacher_data(teachers, src, src_mask, trg_input, trg_mask,
                     noise_std, channel_type, temperature, pad_idx,
                     kd_method, true_labels=None, trim_ratio=0.2):
    channels = Channels()
    logits_list = []
    
    # Feature lists for Feature KD
    feat_x_list = [] # Channel encoded
    
    with torch.no_grad():
        for teacher in teachers:
            enc = teacher.encoder(src, src_mask)
            # Feature KD: Channel Encoded (x)
            tx = PowerNormalize(teacher.channel_encoder(enc))
            feat_x_list.append(tx)

            if channel_type == "AWGN":
                rx = channels.AWGN(tx, noise_std)
            elif channel_type == "Rayleigh":
                rx = channels.Rayleigh(tx, noise_std)
            else:
                rx = channels.Rician(tx, noise_std)

            memory = teacher.channel_decoder(rx)
            dec_out = teacher.decoder(trg_input, memory, trg_mask, None)
            logits = teacher.dense(dec_out)
            logits_list.append(logits)

    stacked_logits = torch.stack(logits_list, dim=0)
    stacked_x = torch.stack(feat_x_list, dim=0)

    fusion_methods = {
        'hinton': fusion_hinton,
        'cwa': fusion_cwa,
        'ueed': fusion_ueed,
        'robust_mean': fusion_robust_mean,
        'robust_median': fusion_robust_median,
        'robust_trimmed': fusion_robust_trimmed_mean,
        'robust_krum': fusion_robust_krum,
    }

    fusion_fn = fusion_methods[kd_method]
    
    # 1. Generate Soft Targets (Logits) - uses decoder output (Length = Target Len)
    soft_targets = fusion_fn(
        stacked_logits, temperature,
        true_labels=true_labels,
        pad_idx=pad_idx,
        trim_ratio=trim_ratio
    )
    
    # 2. Generate Weights for Features (Encoder Output - Length = Source Len)
    # CRITICAL FIX: We cannot use decoder confidence (logits) to weight encoder features 
    # because the lengths differ (Src=31 vs Trg=30).
    # For feature distillation on 'x', we default to uniform weights across teachers
    # to avoid dimension mismatch errors.
    
    K = stacked_logits.size(0)
    fusion_weights = torch.ones(K, 1, 1, 1, device=stacked_logits.device) / K

    return {
        'soft_targets': soft_targets,
        'x': stacked_x,
        'weights': fusion_weights
    }

def kd_train_step(student, teachers, src, trg, noise_std, pad_idx, optimizer,
                  criterion, channel_type, temperature, alpha, kd_method, trim_ratio, feature_weight=0.0):
    optimizer.zero_grad()

    src_mask = (src == pad_idx).unsqueeze(-2).type(torch.float).to(device)
    trg_inp = trg[:, :-1]
    trg_real = trg[:, 1:]

    trg_inp_mask = (trg_inp == pad_idx).unsqueeze(-2).type(torch.float).to(device)
    look_ahead = subsequent_mask(trg_inp.size(1)).type(torch.float).to(device)
    combined_mask_inp = torch.max(trg_inp_mask, look_ahead)

    # Get Teacher Data
    teacher_data = get_teacher_data(
        teachers, src, src_mask, trg_inp, combined_mask_inp,
        noise_std, channel_type, temperature, pad_idx, kd_method,
        true_labels=trg_real, trim_ratio=trim_ratio
    )
    soft_targets = teacher_data['soft_targets']
    
    # Student forward pass
    channels = Channels()
    enc_output = student.encoder(src, src_mask)
    tx_sig = PowerNormalize(student.channel_encoder(enc_output))

    if channel_type == "AWGN":
        rx_sig = channels.AWGN(tx_sig, noise_std)
    elif channel_type == "Rayleigh":
        rx_sig = channels.Rayleigh(tx_sig, noise_std)
    else:
        rx_sig = channels.Rician(tx_sig, noise_std)

    channel_dec_output = student.channel_decoder(rx_sig)
    dec_output = student.decoder(trg_inp, channel_dec_output, combined_mask_inp, None)
    student_logits = student.dense(dec_output)

    # Mask for valid tokens
    mask = (trg_real != pad_idx).float().reshape(-1)
    ntokens = mask.sum() + 1e-8

    # 1. Hard Loss (Cross Entropy)
    # Using ignore_index in criterion is safer than manual masking if reduction is 'none' but followed by mean.
    # Here we do manual to be explicit with ntokens.
    loss_hard_raw = criterion(student_logits.reshape(-1, student_logits.size(-1)), trg_real.reshape(-1))
    loss_hard = (loss_hard_raw * mask).sum() / ntokens

    # 2. Soft Loss (KL Divergence on Logits)
    student_probs_temp = torch.log_softmax(student_logits / temperature, dim=-1) # [B, L, V]
    kl_loss_fn = nn.KLDivLoss(reduction='none')
    loss_soft_raw = kl_loss_fn(student_probs_temp, soft_targets).sum(dim=-1) # [B, L]
    
    # Mask soft loss
    mask_soft = (trg_real != pad_idx).float()
    loss_soft = (loss_soft_raw * mask_soft).sum() / ntokens

    # 3. Feature Loss (KL on Softmaxed Features)
    # Paper Eq 26: KL(Q(O_S)||Q(O_T)) where Q is softmax with temperature.
    loss_feature = torch.tensor(0.0, device=device)
    if feature_weight > 0:
        teacher_x = teacher_data['x'] # [K, B, L, D]
        # student tx_sig: [B, L, D]
        
        if teacher_x.shape[-1] == tx_sig.shape[-1]:
            # Aggregate teacher features distributions
            weights = teacher_data['weights'] # [K, B, L, 1]
            
            # Apply softmax to features (as per paper)
            teacher_probs = torch.softmax(teacher_x / temperature, dim=-1) # [K, B, L, D]
            
            # Weighted average of teacher distributions
            # weights is broadcast to [K, B, L, D]
            avg_teacher_probs = (teacher_probs * weights).sum(dim=0) # [B, L, D]
            
            # Student log probabilities
            student_log_probs = torch.log_softmax(tx_sig / temperature, dim=-1) # [B, L, D]
            
            # KL Divergence
            loss_x_raw = nn.KLDivLoss(reduction='none')(student_log_probs, avg_teacher_probs).sum(dim=-1) # [B, L]
            
            # Mask (using src mask since 'x' is encoder output)
            mask_src = (src != pad_idx).float()
            loss_feature = (loss_x_raw * mask_src).sum() / (mask_src.sum() + 1e-8)

    loss = (1 - alpha) * loss_hard + alpha * (temperature ** 2) * loss_soft + feature_weight * loss_feature

    loss.backward()
    torch.nn.utils.clip_grad_norm_(student.parameters(), 0.5)
    optimizer.step()

    return loss.item(), loss_hard.item(), loss_soft.item()

# ===================================================================
# TRAINING FUNCTION
# ===================================================================

def train_student_with_method(method_config, teachers, vocab, config):
    method_name = method_config['name']

    print(f"\\n{'='*70}")
    print(f"Training with: {method_config['description']}")
    print(f"Paper: {method_config['paper']}")
    print(f"{'='*70}")

    setup_seed(config["seed"])

    token_to_idx = vocab['token_to_idx']
    num_vocab = len(token_to_idx)
    pad_idx = token_to_idx["<PAD>"]

    # Initialize student
    student = DeepSC(
        config["student_num_layers"], num_vocab, num_vocab,
        num_vocab, num_vocab, config["student_d_model"],
        config["student_num_heads"], config["student_dff"], 0.1
    ).to(device)
    initNetParams(student)

    teacher_params = sum(p.numel() for p in teachers[0].parameters())
    student_params = sum(p.numel() for p in student.parameters())
    print(f"Compression: {teacher_params/student_params:.2f}x\\n")

    # Use ignore_index to avoid padding loss in raw criterion calculation
    criterion = nn.CrossEntropyLoss(reduction='none', ignore_index=pad_idx)
    
    # TUNED: Lower LR slightly and add Scheduler
    optimizer = torch.optim.Adam(student.parameters(),
                                 lr=1e-4, betas=(0.9, 0.98), eps=1e-8, weight_decay=5e-4)
    
    # Scheduler to handle the "cup shape" / overfitting
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
        optimizer, mode='min', factor=0.5, patience=3, verbose=True
    )

    train_eur = EurDataset(split='train')
    train_loader = DataLoader(train_eur, batch_size=config["batch_size"],
                              num_workers=0, pin_memory=True, collate_fn=collate_data)

    test_eur = EurDataset('test')
    test_loader = DataLoader(test_eur, batch_size=config["batch_size"],
                             num_workers=0, pin_memory=True, collate_fn=collate_data)

    best_val_loss = float('inf')
    best_model_dict = None
    trim_ratio = method_config['params'].get('robust_trim_ratio', 0.2)
    feature_w = config.get('feature_loss_weight', 0.0)
    
    # TUNED: Use higher alpha if specified in config, else default
    # High alpha (0.9) trusts teacher more, reducing overfitting to transfer set
    alpha = config.get("alpha", 0.9) 

    for epoch in range(config["epochs"]):
        # Training
        student.train()
        pbar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{config['epochs']} [TRAIN]")
        noise_std = np.random.uniform(SNR_to_noise(5), SNR_to_noise(10), size=(1))[0]

        for sents in pbar:
            sents = sents.to(device)
            loss, loss_hard, loss_soft = kd_train_step(
                student, teachers, sents, sents, noise_std, pad_idx,
                optimizer, criterion, config["channel"], config["temperature"],
                alpha, method_name, trim_ratio, feature_w
            )
            pbar.set_postfix({'loss': f'{loss:.4f}'})

        # Validation
        student.eval()
        val_losses = []
        with torch.no_grad():
            for sents in test_loader:
                sents = sents.to(device)
                src_mask = (sents == pad_idx).unsqueeze(-2).type(torch.float).to(device)
                trg_inp = sents[:, :-1]
                trg_real = sents[:, 1:]

                trg_inp_mask = (trg_inp == pad_idx).unsqueeze(-2).type(torch.float).to(device)
                look_ahead = subsequent_mask(trg_inp.size(1)).type(torch.float).to(device)
                combined_mask_inp = torch.max(trg_inp_mask, look_ahead)

                teacher_data = get_teacher_data(
                    teachers, sents, src_mask, trg_inp, combined_mask_inp,
                    SNR_to_noise(8), config["channel"], config["temperature"], pad_idx,
                    method_name, true_labels=trg_real, trim_ratio=trim_ratio
                )
                soft_targets = teacher_data['soft_targets']

                channels = Channels()
                enc_output = student.encoder(sents, src_mask)
                tx_sig = PowerNormalize(student.channel_encoder(enc_output))

                if config["channel"] == "AWGN":
                    rx_sig = channels.AWGN(tx_sig, SNR_to_noise(8))
                elif config["channel"] == "Rayleigh":
                    rx_sig = channels.Rayleigh(tx_sig, SNR_to_noise(8))
                else:
                    rx_sig = channels.Rician(tx_sig, SNR_to_noise(8))

                channel_dec_output = student.channel_decoder(rx_sig)
                dec_output = student.decoder(trg_inp, channel_dec_output, combined_mask_inp, None)
                student_logits = student.dense(dec_output)

                mask = (trg_real != pad_idx).float().reshape(-1)
                ntokens = mask.sum() + 1e-8
                
                loss_hard_raw = criterion(student_logits.reshape(-1, student_logits.size(-1)), trg_real.reshape(-1))
                loss_hard = (loss_hard_raw * mask).sum() / ntokens

                student_probs_temp = torch.log_softmax(student_logits / config["temperature"], dim=-1)
                kl_loss_fn = nn.KLDivLoss(reduction='none')
                loss_soft_raw = kl_loss_fn(student_probs_temp, soft_targets).sum(dim=-1)
                mask_soft = (trg_real != pad_idx).float()
                loss_soft = (loss_soft_raw * mask_soft).sum() / ntokens

                loss = (1 - alpha) * loss_hard + alpha * (config["temperature"] ** 2) * loss_soft
                val_losses.append(loss.item())

        val_loss = np.mean(val_losses)
        
        # Step Scheduler
        scheduler.step(val_loss)
        
        print(f"Epoch {epoch+1}: Val Loss = {val_loss:.5f}")

        if val_loss < best_val_loss:
            best_val_loss = val_loss
            best_model_dict = copy.deepcopy(student.state_dict())
            print(f"\\u2713 New best model!")

    # Save
    ckpt_dir = f"checkpoints/kd_comparison/{method_name}"
    os.makedirs(ckpt_dir, exist_ok=True)
    ckpt_path = f"{ckpt_dir}/best_student_{method_name}.pth"
    torch.save(best_model_dict, ckpt_path)
    print(f"\\u2713 Training completed: {method_name}")
    print(f"\\u2713 Saved: {ckpt_path}\\n")

    return {"status": "success", "ckpt_path": ckpt_path, "method": method_name}

# ===================================================================
# EVALUATION FUNCTION
# ===================================================================

@torch.no_grad()
def decode_greedy(model, src, noise_var, max_len, channel_type, pad_idx, start_idx):
    channels = Channels()

    src_mask = (src == pad_idx).unsqueeze(-2).float().to(device)
    enc = model.encoder(src, src_mask)
    tx = PowerNormalize(model.channel_encoder(enc))

    if channel_type == "AWGN":
        rx = channels.AWGN(tx, noise_var)
    elif channel_type == "Rayleigh":
        rx = channels.Rayleigh(tx, noise_var)
    else:
        rx = channels.Rician(tx, noise_var)

    memory = model.channel_decoder(rx)
    outputs = torch.ones(src.size(0), 1).fill_(start_idx).type_as(src)

    for _ in range(max_len - 1):
        trg_mask = (outputs == pad_idx).unsqueeze(-2).float()
        look_ahead = subsequent_mask(outputs.size(1)).type_as(trg_mask)
        mask = torch.max(trg_mask, look_ahead).to(device)

        dec = model.decoder(outputs, memory, mask, None)
        logits = model.dense(dec)[:, -1:, :]
        next_word = torch.argmax(logits, dim=-1)
        outputs = torch.cat([outputs, next_word], dim=1)

    return outputs

def evaluate_method(method_name, ckpt_path, vocab, config):
    print(f"\\nEvaluating {method_name}...")

    token_to_idx = vocab["token_to_idx"]
    pad_idx = token_to_idx["<PAD>"]
    start_idx = token_to_idx["<START>"]
    end_idx = token_to_idx["<END>"]

    # Load student
    student_cfg = {
        "num_layers": config["student_num_layers"],
        "src_vocab_size": len(token_to_idx),
        "trg_vocab_size": len(token_to_idx),
        "src_max_len": len(token_to_idx),
        "trg_max_len": len(token_to_idx),
        "d_model": config["student_d_model"],
        "num_heads": config["student_num_heads"],
        "dff": config["student_dff"],
        "dropout": 0.1,
    }

    student = DeepSC(**student_cfg).to(device)
    state = torch.load(ckpt_path, map_location=device)
    student.load_state_dict(state)
    student.eval()

    dataset = EurDataset(split="test")
    loader = DataLoader(dataset, batch_size=32, num_workers=0,
                        pin_memory=True, collate_fn=collate_data)

    bleu = BleuScore(1, 0, 0, 0)
    sbert = SentenceTransformer("all-MiniLM-L6-v2")
    seq2text = SeqtoText(token_to_idx, end_idx)

    results = {"snr": [0, 6, 12, 18], "bleu": [], "sim": []}

    for snr in results["snr"]:
        noise = SNR_to_noise(snr)
        tx, rx = [], []

        for batch in tqdm(loader, desc=f"SNR {snr}dB"):
            batch = batch.to(device)
            preds = decode_greedy(student, batch, noise, 30,
                                 config["channel"], pad_idx, start_idx)

            tx.extend(map(seq2text.sequence_to_text, preds.cpu().numpy().tolist()))
            rx.extend(map(seq2text.sequence_to_text, batch.cpu().numpy().tolist()))

        bleu_scores = bleu.compute_blue_score(rx, tx)
        sim_scores = sbert.encode(rx, convert_to_tensor=True)
        pred_embed = sbert.encode(tx, convert_to_tensor=True)
        sim_val = util.cos_sim(sim_scores, pred_embed).diagonal().cpu().numpy()

        results["bleu"].append(float(np.mean(bleu_scores)))
        results["sim"].append(float(np.mean(sim_val)))

        print(f"SNR {snr}dB -> BLEU: {results['bleu'][-1]:.4f}, Sim: {results['sim'][-1]:.4f}")

    # Save per-method results
    output_dir = f"results/kd_comparison/{method_name}"
    os.makedirs(output_dir, exist_ok=True)
    out_file = Path(output_dir) / "student_model_results.json"
    with open(out_file, "w", encoding="utf-8") as f:
        json.dump(results, f, indent=2)

    avg_bleu = sum(results["bleu"]) / len(results["bleu"])
    avg_sim = sum(results["sim"]) / len(results["sim"])

    print(f"\\u2713 {method_name}: BLEU={avg_bleu:.4f}, Sim={avg_sim:.4f}")
    return {"status": "success", "results": results, "avg_bleu": avg_bleu, "avg_sim": avg_sim}

# ===================================================================
# COMPARISON & VISUALIZATION
# ===================================================================

def create_comparison_plots(df):
    if "avg_bleu" not in df.columns:
        return

    fig, axes = plt.subplots(1, 2, figsize=(14, 5))

    # Bar chart
    df_sorted = df.sort_values("avg_bleu", ascending=False)
    axes[0].barh(df_sorted["method"], df_sorted["avg_bleu"])
    axes[0].set_xlabel("Average BLEU Score")
    axes[0].set_title("KD Methods Comparison - Average BLEU")
    axes[0].grid(axis='x', alpha=0.3)

    # Line plot
    snr_cols = [col for col in df.columns if col.startswith("bleu_snr")]
    snrs = [int(col.replace("bleu_snr", "")) for col in snr_cols]

    for _, row in df.iterrows():
        bleus = [row[col] for col in snr_cols]
        axes[1].plot(snrs, bleus, marker='o', label=row["method"])

    axes[1].set_xlabel("SNR (dB)")
    axes[1].set_ylabel("BLEU Score")
    axes[1].set_title("BLEU vs SNR - All Methods")
    axes[1].legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    axes[1].grid(alpha=0.3)

    plt.tight_layout()
    plt.savefig("results/kd_comparison/methods_comparison.png", dpi=150, bbox_inches='tight')
    print(f"\\n\\ud83d\\udcca Plots saved to: results/kd_comparison/methods_comparison.png")
    plt.show()

# ===================================================================
# MAIN EXECUTION
# ===================================================================

print(f"\\n{'='*70}")
print(f"KD Methods Comparison")
print(f"Mode: train_and_eval")
print(f"Methods: {len(KD_METHODS)}")
print(f"{'='*70}\\n")

print("Loading vocab...")
vocab = json.load(open(CONFIG["vocab_file"], 'rb'))
print(f"\\u2713 Vocab: {len(vocab['token_to_idx'])} tokens\\n")

print("Loading teacher checkpoints...")
teacher_ckpts = list_teacher_checkpoints(CONFIG["teacher_ckpt_root"], CONFIG["k_teachers"])
print(f"\\u2713 Found {len(teacher_ckpts)} checkpoints")
for i, p in enumerate(teacher_ckpts):
    print(f"  {i+1}. {p}")
print()

print("Loading teacher models...")
teachers = load_teacher_models(teacher_ckpts, len(vocab["token_to_idx"]), CONFIG)
print(f"\\u2713 Teachers loaded\\n")

# Run comparison
all_results = []

for method_config in KD_METHODS:
    method_result = {
        "method": method_config['name'],
        "description": method_config['description'],
        "paper": method_config['paper'],
    }

    # Training
    try:
        train_result = train_student_with_method(method_config, teachers, vocab, CONFIG)
        method_result["train_status"] = train_result["status"]
        ckpt_path = train_result["ckpt_path"]
    except Exception as e:
        print(f"\\u2717 Training failed: {method_config['name']}")
        print(f"Error: {e}")
        method_result["train_status"] = "failed"
        all_results.append(method_result)
        continue

    # Evaluation
    try:
        eval_result = evaluate_method(method_config['name'], ckpt_path, vocab, CONFIG)
        method_result["eval_status"] = eval_result["status"]
        method_result["avg_bleu"] = eval_result["avg_bleu"]
        method_result["avg_sim"] = eval_result["avg_sim"]

        results = eval_result["results"]
        for i, snr in enumerate(results["snr"]):
            method_result[f"bleu_snr{snr}"] = results["bleu"][i]
            method_result[f"sim_snr{snr}"] = results["sim"][i]
    except Exception as e:
        print(f"\\u2717 Evaluation failed: {method_config['name']}")
        print(f"Error: {e}")
        method_result["eval_status"] = "failed"

    all_results.append(method_result)

# Save results
os.makedirs("results/kd_comparison", exist_ok=True)
df = pd.DataFrame(all_results)

summary_file = "results/kd_comparison/methods_comparison.csv"
df.to_csv(summary_file, index=False)

print(f"\\n{'='*70}")
print(f"Comparison complete!")
print(f"Results saved to: {summary_file}")
print(f"{'='*70}\\n")

# Display results
if "avg_bleu" in df.columns:
    print("\\nResults Summary (sorted by Avg BLEU):")
    print("=" * 90)
    display_cols = ["method", "description", "avg_bleu", "avg_sim"]
    display_df = df[display_cols].sort_values("avg_bleu", ascending=False)
    print(display_df.to_string(index=False))

    # Best method
    best_method = df.loc[df["avg_bleu"].idxmax()]
    print(f"\\n\\ud83c\\udfc6 Best Method: {best_method['method']}")
    print(f"   Description: {best_method['description']}")
    print(f"   Avg BLEU: {best_method['avg_bleu']:.4f}")
    print(f"   Paper: {best_method['paper']}")

    # Create visualization
    create_comparison_plots(df)

print("\\n\\u2713 COMPLETE!")



```

```python

# ===================================================================
# COMPLETE ALL-IN-ONE KD COMPARISON - FIXED & ENHANCED
# No subprocess, includes training, evaluation, CSV, and plots
# ===================================================================

import os
import json
import torch
import torch.nn as nn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
from tqdm.notebook import tqdm
import copy
import random
import math
from sentence_transformers import SentenceTransformer, util

# Your imports
from dataset import EurDataset, collate_data
from models.transceiver import DeepSC
from utils import BleuScore, SNR_to_noise, SeqtoText, initNetParams, PowerNormalize, Channels, subsequent_mask
from torch.utils.data import DataLoader

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# ===================================================================
# CONFIGURATION
# ===================================================================

CONFIG = {
    "vocab_file": "../europarl/vocab.json",
    "teacher_ckpt_root": "../europarl/checkpoints/deepsc-Rayleigh/k5_jammoul/rayleigh_channel/",
    "channel": "Rayleigh",
    "epochs": 40,
    "batch_size": 64,
    "k_teachers": 5,
    "temperature": 2.0,
    "alpha": 0.9, # TUNED: High alpha to trust teacher and prevent overfitting
    "feature_loss_weight": 0.1,
    "seed": 42,
    "student_d_model": 128,
    "student_dff": 256,
    "student_num_layers": 2,
    "student_num_heads": 8,
    "teacher_d_model": 128,
    "teacher_dff": 512,
    "teacher_num_layers": 4,
    "teacher_num_heads": 8,
}

KD_METHODS = [
    {
        'name': 'hinton',
        'description': 'Hinton Temperature Distillation (uniform averaging)',
        'paper': 'Hinton et al. 2015',
        'params': {}
    },
    {
        'name': 'cwa',
        'description': 'Confidence-Weighted Averaging (your method)',
        'paper': 'Your ensemble work',
        'params': {}
    },
    {
        'name': 'ueed',
        'description': 'UEED - Correctness-weighted teachers',
        'paper': 'Wu et al. 2022',
        'params': {}
    },
    {
        'name': 'robust_mean',
        'description': 'Robust Aggregation - Mean of logits',
        'paper': 'Hamdi et al. 2025',
        'params': {}
    },
    {
        'name': 'robust_median',
        'description': 'Robust Aggregation - Median',
        'paper': 'Hamdi et al. 2025',
        'params': {}
    },
    {
        'name': 'robust_trimmed',
        'description': 'Robust Aggregation - Trimmed mean',
        'paper': 'Hamdi et al. 2025',
        'params': {'robust_trim_ratio': 0.2}
    },
    {
        'name': 'robust_krum',
        'description': 'Robust Aggregation - Krum (consensus)',
        'paper': 'Hamdi et al. 2025',
        'params': {}
    },
]

# ===================================================================
# HELPER FUNCTIONS
# ===================================================================

def setup_seed(seed):
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    np.random.seed(seed)
    random.seed(seed)
    torch.backends.cudnn.deterministic = True

def list_teacher_checkpoints(base_dir, k):
    paths = []
    for seed_dir in sorted(Path(base_dir).glob("seed*")):
        ckpts = sorted(seed_dir.glob("checkpoint_*.pth"))
        if ckpts:
            paths.append(str(ckpts[-1]))
    if len(paths) < k:
        raise RuntimeError(f"Found only {len(paths)} checkpoints under {base_dir}, need {k}")
    return paths[:k]

def load_teacher_models(ckpt_paths, num_vocab, config):
    models = []
    for i, p in enumerate(ckpt_paths):
        print(f"Loading teacher {i+1}/{len(ckpt_paths)}...", flush=True)
        model = DeepSC(
            config["teacher_num_layers"], num_vocab, num_vocab,
            num_vocab, num_vocab, config["teacher_d_model"],
            config["teacher_num_heads"], config["teacher_dff"], 0.1
        ).to(device)
        state = torch.load(p, map_location=device)
        model.load_state_dict(state)
        model.eval()
        models.append(model)
    return models

# ===================================================================
# FUSION METHODS
# ===================================================================

def fusion_hinton(stacked_logits, temperature, **kwargs):
    probs_temp = torch.softmax(stacked_logits / temperature, dim=-1)
    return probs_temp.mean(dim=0)

def fusion_cwa(stacked_logits, temperature, **kwargs):
    probs = torch.softmax(stacked_logits, dim=-1)
    ent = -(probs * torch.log(probs + 1e-10)).sum(dim=-1)
    norm_ent = ent / math.log(probs.size(-1))
    confidences = 1 - norm_ent
    weights = confidences / (confidences.sum(dim=0, keepdim=True) + 1e-10)
    weights = weights.unsqueeze(-1)
    probs_temp = torch.softmax(stacked_logits / temperature, dim=-1)
    return (weights * probs_temp).sum(dim=0)

def fusion_ueed(stacked_logits, temperature, true_labels=None, pad_idx=None, **kwargs):
    probs = torch.softmax(stacked_logits, dim=-1)
    if true_labels is not None:
        teacher_preds = torch.argmax(probs, dim=-1)
        true_labels_exp = true_labels.unsqueeze(0).expand(probs.size(0), -1, -1)
        correct = (teacher_preds == true_labels_exp).float()
        if pad_idx is not None:
            pad_mask = (true_labels != pad_idx).float()
            correct = correct * pad_mask.unsqueeze(0)
        weights = correct / (correct.sum(dim=0, keepdim=True) + 1e-10)
        weights = weights.unsqueeze(-1)
        probs_temp = torch.softmax(stacked_logits / temperature, dim=-1)
        return (weights * probs_temp).sum(dim=0)
    else:
        probs_temp = torch.softmax(stacked_logits / temperature, dim=-1)
        return probs_temp.mean(dim=0)

def fusion_robust_median(stacked_logits, temperature, **kwargs):
    median_logits = torch.median(stacked_logits, dim=0).values
    return torch.softmax(median_logits / temperature, dim=-1)

def fusion_robust_trimmed_mean(stacked_logits, temperature, trim_ratio=0.2, **kwargs):
    K = stacked_logits.size(0)
    k_trim = max(1, int(K * trim_ratio))
    sorted_logits, _ = torch.sort(stacked_logits, dim=0)
    trimmed_logits = sorted_logits[k_trim:-k_trim] if k_trim > 0 else sorted_logits
    mean_logits = trimmed_logits.mean(dim=0)
    return torch.softmax(mean_logits / temperature, dim=-1)

def fusion_robust_krum(stacked_logits, temperature, **kwargs):
    K = stacked_logits.shape[0]
    flat_logits = stacked_logits.reshape(K, -1)
    distances = torch.zeros(K, device=stacked_logits.device)
    for k in range(K):
        for j in range(K):
            if j != k:
                distances[k] += torch.sum((flat_logits[k] - flat_logits[j]) ** 2)
    best_teacher_idx = torch.argmin(distances)
    consensus_logits = stacked_logits[best_teacher_idx]
    return torch.softmax(consensus_logits / temperature, dim=-1)

def fusion_robust_mean(stacked_logits, temperature, **kwargs):
    mean_logits = stacked_logits.mean(dim=0)
    return torch.softmax(mean_logits / temperature, dim=-1)

# ===================================================================
# UNIFIED TARGET GENERATION & DATA RETRIEVAL
# ===================================================================

def get_teacher_data(teachers, src, src_mask, trg_input, trg_mask,
                     noise_std, channel_type, temperature, pad_idx,
                     kd_method, true_labels=None, trim_ratio=0.2):
    channels = Channels()
    logits_list = []
    
    # Feature lists for Feature KD
    feat_x_list = [] # Channel encoded
    
    with torch.no_grad():
        for teacher in teachers:
            enc = teacher.encoder(src, src_mask)
            # Feature KD: Channel Encoded (x)
            tx = PowerNormalize(teacher.channel_encoder(enc))
            feat_x_list.append(tx)

            if channel_type == "AWGN":
                rx = channels.AWGN(tx, noise_std)
            elif channel_type == "Rayleigh":
                rx = channels.Rayleigh(tx, noise_std)
            else:
                rx = channels.Rician(tx, noise_std)

            memory = teacher.channel_decoder(rx)
            dec_out = teacher.decoder(trg_input, memory, trg_mask, None)
            logits = teacher.dense(dec_out)
            logits_list.append(logits)

    stacked_logits = torch.stack(logits_list, dim=0)
    stacked_x = torch.stack(feat_x_list, dim=0)

    fusion_methods = {
        'hinton': fusion_hinton,
        'cwa': fusion_cwa,
        'ueed': fusion_ueed,
        'robust_mean': fusion_robust_mean,
        'robust_median': fusion_robust_median,
        'robust_trimmed': fusion_robust_trimmed_mean,
        'robust_krum': fusion_robust_krum,
    }

    fusion_fn = fusion_methods[kd_method]
    
    # 1. Generate Soft Targets (Logits) - uses decoder output (Length = Target Len)
    soft_targets = fusion_fn(
        stacked_logits, temperature,
        true_labels=true_labels,
        pad_idx=pad_idx,
        trim_ratio=trim_ratio
    )
    
    # 2. Generate Weights for Features (Encoder Output - Length = Source Len)
    # CRITICAL FIX: We cannot use decoder confidence (logits) to weight encoder features 
    # because the lengths differ (Src=31 vs Trg=30).
    # For feature distillation on 'x', we default to uniform weights across teachers
    # to avoid dimension mismatch errors.
    
    K = stacked_logits.size(0)
    fusion_weights = torch.ones(K, 1, 1, 1, device=stacked_logits.device) / K

    return {
        'soft_targets': soft_targets,
        'x': stacked_x,
        'weights': fusion_weights
    }

def kd_train_step(student, teachers, src, trg, noise_std, pad_idx, optimizer,
                  criterion, channel_type, temperature, alpha, kd_method, trim_ratio, feature_weight=0.0):
    optimizer.zero_grad()

    src_mask = (src == pad_idx).unsqueeze(-2).type(torch.float).to(device)
    trg_inp = trg[:, :-1]
    trg_real = trg[:, 1:]

    trg_inp_mask = (trg_inp == pad_idx).unsqueeze(-2).type(torch.float).to(device)
    look_ahead = subsequent_mask(trg_inp.size(1)).type(torch.float).to(device)
    combined_mask_inp = torch.max(trg_inp_mask, look_ahead)

    # Get Teacher Data
    teacher_data = get_teacher_data(
        teachers, src, src_mask, trg_inp, combined_mask_inp,
        noise_std, channel_type, temperature, pad_idx, kd_method,
        true_labels=trg_real, trim_ratio=trim_ratio
    )
    soft_targets = teacher_data['soft_targets']
    
    # Student forward pass
    channels = Channels()
    enc_output = student.encoder(src, src_mask)
    tx_sig = PowerNormalize(student.channel_encoder(enc_output))

    if channel_type == "AWGN":
        rx_sig = channels.AWGN(tx_sig, noise_std)
    elif channel_type == "Rayleigh":
        rx_sig = channels.Rayleigh(tx_sig, noise_std)
    else:
        rx_sig = channels.Rician(tx_sig, noise_std)

    channel_dec_output = student.channel_decoder(rx_sig)
    dec_output = student.decoder(trg_inp, channel_dec_output, combined_mask_inp, None)
    student_logits = student.dense(dec_output)

    # Mask for valid tokens
    mask = (trg_real != pad_idx).float().reshape(-1)
    ntokens = mask.sum() + 1e-8

    # 1. Hard Loss (Cross Entropy)
    # Using ignore_index in criterion is safer than manual masking if reduction is 'none' but followed by mean.
    # Here we do manual to be explicit with ntokens.
    loss_hard_raw = criterion(student_logits.reshape(-1, student_logits.size(-1)), trg_real.reshape(-1))
    loss_hard = (loss_hard_raw * mask).sum() / ntokens

    # 2. Soft Loss (KL Divergence on Logits)
    student_probs_temp = torch.log_softmax(student_logits / temperature, dim=-1) # [B, L, V]
    kl_loss_fn = nn.KLDivLoss(reduction='none')
    loss_soft_raw = kl_loss_fn(student_probs_temp, soft_targets).sum(dim=-1) # [B, L]
    
    # Mask soft loss
    mask_soft = (trg_real != pad_idx).float()
    loss_soft = (loss_soft_raw * mask_soft).sum() / ntokens

    # 3. Feature Loss (KL on Softmaxed Features)
    # Paper Eq 26: KL(Q(O_S)||Q(O_T)) where Q is softmax with temperature.
    loss_feature = torch.tensor(0.0, device=device)
    if feature_weight > 0:
        teacher_x = teacher_data['x'] # [K, B, L, D]
        # student tx_sig: [B, L, D]
        
        if teacher_x.shape[-1] == tx_sig.shape[-1]:
            # Aggregate teacher features distributions
            weights = teacher_data['weights'] # [K, B, L, 1]
            
            # Apply softmax to features (as per paper)
            teacher_probs = torch.softmax(teacher_x / temperature, dim=-1) # [K, B, L, D]
            
            # Weighted average of teacher distributions
            # weights is broadcast to [K, B, L, D]
            avg_teacher_probs = (teacher_probs * weights).sum(dim=0) # [B, L, D]
            
            # Student log probabilities
            student_log_probs = torch.log_softmax(tx_sig / temperature, dim=-1) # [B, L, D]
            
            # KL Divergence
            loss_x_raw = nn.KLDivLoss(reduction='none')(student_log_probs, avg_teacher_probs).sum(dim=-1) # [B, L]
            
            # Mask (using src mask since 'x' is encoder output)
            mask_src = (src != pad_idx).float()
            loss_feature = (loss_x_raw * mask_src).sum() / (mask_src.sum() + 1e-8)

    loss = (1 - alpha) * loss_hard + alpha * (temperature ** 2) * loss_soft + feature_weight * loss_feature

    loss.backward()
    torch.nn.utils.clip_grad_norm_(student.parameters(), 0.5)
    optimizer.step()

    return loss.item(), loss_hard.item(), loss_soft.item()

# ===================================================================
# TRAINING FUNCTION
# ===================================================================

def train_student_with_method(method_config, teachers, vocab, config):
    method_name = method_config['name']

    print(f"\\n{'='*70}")
    print(f"Training with: {method_config['description']}")
    print(f"Paper: {method_config['paper']}")
    print(f"{'='*70}")

    setup_seed(config["seed"])

    token_to_idx = vocab['token_to_idx']
    num_vocab = len(token_to_idx)
    pad_idx = token_to_idx["<PAD>"]

    # Initialize student
    student = DeepSC(
        config["student_num_layers"], num_vocab, num_vocab,
        num_vocab, num_vocab, config["student_d_model"],
        config["student_num_heads"], config["student_dff"], 0.1
    ).to(device)
    initNetParams(student)

    teacher_params = sum(p.numel() for p in teachers[0].parameters())
    student_params = sum(p.numel() for p in student.parameters())
    print(f"Compression: {teacher_params/student_params:.2f}x\\n")

    # Use ignore_index to avoid padding loss in raw criterion calculation
    criterion = nn.CrossEntropyLoss(reduction='none', ignore_index=pad_idx)
    
    optimizer = torch.optim.Adam(student.parameters(),
                                 lr=5e-4, betas=(0.9, 0.98), eps=1e-8, weight_decay=5e-4)

    train_eur = EurDataset(split='train')
    train_loader = DataLoader(train_eur, batch_size=config["batch_size"],
                              num_workers=0, pin_memory=True, collate_fn=collate_data)

    test_eur = EurDataset('test')
    test_loader = DataLoader(test_eur, batch_size=config["batch_size"],
                             num_workers=0, pin_memory=True, collate_fn=collate_data)

    best_val_loss = float('inf')
    best_model_dict = None
    trim_ratio = method_config['params'].get('robust_trim_ratio', 0.2)
    feature_w = config.get('feature_loss_weight', 0.0)

    for epoch in range(config["epochs"]):
        # Training
        student.train()
        pbar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{config['epochs']} [TRAIN]")
        noise_std = np.random.uniform(SNR_to_noise(5), SNR_to_noise(10), size=(1))[0]

        for sents in pbar:
            sents = sents.to(device)
            loss, loss_hard, loss_soft = kd_train_step(
                student, teachers, sents, sents, noise_std, pad_idx,
                optimizer, criterion, config["channel"], config["temperature"],
                config["alpha"], method_name, trim_ratio, feature_w
            )
            pbar.set_postfix({'loss': f'{loss:.4f}'})

        # Validation
        student.eval()
        val_losses = []
        with torch.no_grad():
            for sents in test_loader:
                sents = sents.to(device)
                src_mask = (sents == pad_idx).unsqueeze(-2).type(torch.float).to(device)
                trg_inp = sents[:, :-1]
                trg_real = sents[:, 1:]

                trg_inp_mask = (trg_inp == pad_idx).unsqueeze(-2).type(torch.float).to(device)
                look_ahead = subsequent_mask(trg_inp.size(1)).type(torch.float).to(device)
                combined_mask_inp = torch.max(trg_inp_mask, look_ahead)

                teacher_data = get_teacher_data(
                    teachers, sents, src_mask, trg_inp, combined_mask_inp,
                    SNR_to_noise(8), config["channel"], config["temperature"], pad_idx,
                    method_name, true_labels=trg_real, trim_ratio=trim_ratio
                )
                soft_targets = teacher_data['soft_targets']

                channels = Channels()
                enc_output = student.encoder(sents, src_mask)
                tx_sig = PowerNormalize(student.channel_encoder(enc_output))

                if config["channel"] == "AWGN":
                    rx_sig = channels.AWGN(tx_sig, SNR_to_noise(8))
                elif config["channel"] == "Rayleigh":
                    rx_sig = channels.Rayleigh(tx_sig, SNR_to_noise(8))
                else:
                    rx_sig = channels.Rician(tx_sig, SNR_to_noise(8))

                channel_dec_output = student.channel_decoder(rx_sig)
                dec_output = student.decoder(trg_inp, channel_dec_output, combined_mask_inp, None)
                student_logits = student.dense(dec_output)

                mask = (trg_real != pad_idx).float().reshape(-1)
                ntokens = mask.sum() + 1e-8
                
                loss_hard_raw = criterion(student_logits.reshape(-1, student_logits.size(-1)), trg_real.reshape(-1))
                loss_hard = (loss_hard_raw * mask).sum() / ntokens

                student_probs_temp = torch.log_softmax(student_logits / config["temperature"], dim=-1)
                kl_loss_fn = nn.KLDivLoss(reduction='none')
                loss_soft_raw = kl_loss_fn(student_probs_temp, soft_targets).sum(dim=-1)
                mask_soft = (trg_real != pad_idx).float()
                loss_soft = (loss_soft_raw * mask_soft).sum() / ntokens

                loss = (1 - config["alpha"]) * loss_hard + config["alpha"] * (config["temperature"] ** 2) * loss_soft
                val_losses.append(loss.item())

        val_loss = np.mean(val_losses)
        print(f"Epoch {epoch+1}: Val Loss = {val_loss:.5f}")

        if val_loss < best_val_loss:
            best_val_loss = val_loss
            best_model_dict = copy.deepcopy(student.state_dict())
            print(f"\\u2713 New best model!")

    # Save
    ckpt_dir = f"checkpoints/kd_comparison/{method_name}"
    os.makedirs(ckpt_dir, exist_ok=True)
    ckpt_path = f"{ckpt_dir}/best_student_{method_name}.pth"
    torch.save(best_model_dict, ckpt_path)
    print(f"\\u2713 Training completed: {method_name}")
    print(f"\\u2713 Saved: {ckpt_path}\\n")

    return {"status": "success", "ckpt_path": ckpt_path, "method": method_name}

# ===================================================================
# EVALUATION FUNCTION
# ===================================================================

@torch.no_grad()
def decode_greedy(model, src, noise_var, max_len, channel_type, pad_idx, start_idx):
    channels = Channels()

    src_mask = (src == pad_idx).unsqueeze(-2).float().to(device)
    enc = model.encoder(src, src_mask)
    tx = PowerNormalize(model.channel_encoder(enc))

    if channel_type == "AWGN":
        rx = channels.AWGN(tx, noise_var)
    elif channel_type == "Rayleigh":
        rx = channels.Rayleigh(tx, noise_var)
    else:
        rx = channels.Rician(tx, noise_var)

    memory = model.channel_decoder(rx)
    outputs = torch.ones(src.size(0), 1).fill_(start_idx).type_as(src)

    for _ in range(max_len - 1):
        trg_mask = (outputs == pad_idx).unsqueeze(-2).float()
        look_ahead = subsequent_mask(outputs.size(1)).type_as(trg_mask)
        mask = torch.max(trg_mask, look_ahead).to(device)

        dec = model.decoder(outputs, memory, mask, None)
        logits = model.dense(dec)[:, -1:, :]
        next_word = torch.argmax(logits, dim=-1)
        outputs = torch.cat([outputs, next_word], dim=1)

    return outputs

def evaluate_method(method_name, ckpt_path, vocab, config):
    print(f"\\nEvaluating {method_name}...")

    token_to_idx = vocab["token_to_idx"]
    pad_idx = token_to_idx["<PAD>"]
    start_idx = token_to_idx["<START>"]
    end_idx = token_to_idx["<END>"]

    # Load student
    student_cfg = {
        "num_layers": config["student_num_layers"],
        "src_vocab_size": len(token_to_idx),
        "trg_vocab_size": len(token_to_idx),
        "src_max_len": len(token_to_idx),
        "trg_max_len": len(token_to_idx),
        "d_model": config["student_d_model"],
        "num_heads": config["student_num_heads"],
        "dff": config["student_dff"],
        "dropout": 0.1,
    }

    student = DeepSC(**student_cfg).to(device)
    state = torch.load(ckpt_path, map_location=device)
    student.load_state_dict(state)
    student.eval()

    dataset = EurDataset(split="test")
    loader = DataLoader(dataset, batch_size=32, num_workers=0,
                        pin_memory=True, collate_fn=collate_data)

    bleu = BleuScore(1, 0, 0, 0)
    sbert = SentenceTransformer("all-MiniLM-L6-v2")
    seq2text = SeqtoText(token_to_idx, end_idx)

    results = {"snr": [0, 6, 12, 18], "bleu": [], "sim": []}

    for snr in results["snr"]:
        noise = SNR_to_noise(snr)
        tx, rx = [], []

        for batch in tqdm(loader, desc=f"SNR {snr}dB"):
            batch = batch.to(device)
            preds = decode_greedy(student, batch, noise, 30,
                                 config["channel"], pad_idx, start_idx)

            tx.extend(map(seq2text.sequence_to_text, preds.cpu().numpy().tolist()))
            rx.extend(map(seq2text.sequence_to_text, batch.cpu().numpy().tolist()))

        bleu_scores = bleu.compute_blue_score(rx, tx)
        sim_scores = sbert.encode(rx, convert_to_tensor=True)
        pred_embed = sbert.encode(tx, convert_to_tensor=True)
        sim_val = util.cos_sim(sim_scores, pred_embed).diagonal().cpu().numpy()

        results["bleu"].append(float(np.mean(bleu_scores)))
        results["sim"].append(float(np.mean(sim_val)))

        print(f"SNR {snr}dB -> BLEU: {results['bleu'][-1]:.4f}, Sim: {results['sim'][-1]:.4f}")

    # Save per-method results
    output_dir = f"results/kd_comparison/{method_name}"
    os.makedirs(output_dir, exist_ok=True)
    out_file = Path(output_dir) / "student_model_results.json"
    with open(out_file, "w", encoding="utf-8") as f:
        json.dump(results, f, indent=2)

    avg_bleu = sum(results["bleu"]) / len(results["bleu"])
    avg_sim = sum(results["sim"]) / len(results["sim"])

    print(f"\\u2713 {method_name}: BLEU={avg_bleu:.4f}, Sim={avg_sim:.4f}")
    return {"status": "success", "results": results, "avg_bleu": avg_bleu, "avg_sim": avg_sim}

# ===================================================================
# COMPARISON & VISUALIZATION
# ===================================================================

def create_comparison_plots(df):
    if "avg_bleu" not in df.columns:
        return

    fig, axes = plt.subplots(1, 2, figsize=(14, 5))

    # Bar chart
    df_sorted = df.sort_values("avg_bleu", ascending=False)
    axes[0].barh(df_sorted["method"], df_sorted["avg_bleu"])
    axes[0].set_xlabel("Average BLEU Score")
    axes[0].set_title("KD Methods Comparison - Average BLEU")
    axes[0].grid(axis='x', alpha=0.3)

    # Line plot
    snr_cols = [col for col in df.columns if col.startswith("bleu_snr")]
    snrs = [int(col.replace("bleu_snr", "")) for col in snr_cols]

    for _, row in df.iterrows():
        bleus = [row[col] for col in snr_cols]
        axes[1].plot(snrs, bleus, marker='o', label=row["method"])

    axes[1].set_xlabel("SNR (dB)")
    axes[1].set_ylabel("BLEU Score")
    axes[1].set_title("BLEU vs SNR - All Methods")
    axes[1].legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    axes[1].grid(alpha=0.3)

    plt.tight_layout()
    plt.savefig("results/kd_comparison/methods_comparison.png", dpi=150, bbox_inches='tight')
    print(f"\\n\\ud83d\\udcca Plots saved to: results/kd_comparison/methods_comparison.png")
    plt.show()

# ===================================================================
# MAIN EXECUTION
# ===================================================================

print(f"\\n{'='*70}")
print(f"KD Methods Comparison")
print(f"Mode: train_and_eval")
print(f"Methods: {len(KD_METHODS)}")
print(f"{'='*70}\\n")

print("Loading vocab...")
vocab = json.load(open(CONFIG["vocab_file"], 'rb'))
print(f"\\u2713 Vocab: {len(vocab['token_to_idx'])} tokens\\n")

print("Loading teacher checkpoints...")
teacher_ckpts = list_teacher_checkpoints(CONFIG["teacher_ckpt_root"], CONFIG["k_teachers"])
print(f"\\u2713 Found {len(teacher_ckpts)} checkpoints")
for i, p in enumerate(teacher_ckpts):
    print(f"  {i+1}. {p}")
print()

print("Loading teacher models...")
teachers = load_teacher_models(teacher_ckpts, len(vocab["token_to_idx"]), CONFIG)
print(f"\\u2713 Teachers loaded\\n")

# Run comparison
all_results = []

for method_config in KD_METHODS:
    method_result = {
        "method": method_config['name'],
        "description": method_config['description'],
        "paper": method_config['paper'],
    }

    # Training
    try:
        train_result = train_student_with_method(method_config, teachers, vocab, CONFIG)
        method_result["train_status"] = train_result["status"]
        ckpt_path = train_result["ckpt_path"]
    except Exception as e:
        print(f"\\u2717 Training failed: {method_config['name']}")
        print(f"Error: {e}")
        method_result["train_status"] = "failed"
        all_results.append(method_result)
        continue

    # Evaluation
    try:
        eval_result = evaluate_method(method_config['name'], ckpt_path, vocab, CONFIG)
        method_result["eval_status"] = eval_result["status"]
        method_result["avg_bleu"] = eval_result["avg_bleu"]
        method_result["avg_sim"] = eval_result["avg_sim"]

        results = eval_result["results"]
        for i, snr in enumerate(results["snr"]):
            method_result[f"bleu_snr{snr}"] = results["bleu"][i]
            method_result[f"sim_snr{snr}"] = results["sim"][i]
    except Exception as e:
        print(f"\\u2717 Evaluation failed: {method_config['name']}")
        print(f"Error: {e}")
        method_result["eval_status"] = "failed"

    all_results.append(method_result)

# Save results
os.makedirs("results/kd_comparison", exist_ok=True)
df = pd.DataFrame(all_results)

summary_file = "results/kd_comparison/methods_comparison.csv"
df.to_csv(summary_file, index=False)

print(f"\\n{'='*70}")
print(f"Comparison complete!")
print(f"Results saved to: {summary_file}")
print(f"{'='*70}\\n")

# Display results
if "avg_bleu" in df.columns:
    print("\\nResults Summary (sorted by Avg BLEU):")
    print("=" * 90)
    display_cols = ["method", "description", "avg_bleu", "avg_sim"]
    display_df = df[display_cols].sort_values("avg_bleu", ascending=False)
    print(display_df.to_string(index=False))

    # Best method
    best_method = df.loc[df["avg_bleu"].idxmax()]
    print(f"\\n\\ud83c\\udfc6 Best Method: {best_method['method']}")
    print(f"   Description: {best_method['description']}")
    print(f"   Avg BLEU: {best_method['avg_bleu']:.4f}")
    print(f"   Paper: {best_method['paper']}")

    # Create visualization
    create_comparison_plots(df)

print("\\n\\u2713 COMPLETE!")


```
